```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 25
x = np.linspace(0, 2, n)
y = np.linspace(0, 2, n)
xx, yy = np.meshgrid(x,y)
P = xx * yy 
Q = xx**2 * yy**2
```

```python
t = np.linspace(0, np.pi/2, n)
xt = 2*np.cos(t)
yt = 2*np.sin(t)
```

```python
plt.axes([0, 0, 2, 2])
plt.plot(xt, yt, color = 'red')
plt.quiver(xx, yy, P, Q)
plt.show()
```
